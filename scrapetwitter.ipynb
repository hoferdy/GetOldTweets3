```
# import GetOldTweets3
import GetOldTweets3 as got
#set criteria
keyword = 'bsps'

count = 500
```

```
# begin scraping
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(keyword)\
.setMaxTweets(count)
tweets = got.manager.TweetManager.getTweets(tweetCriteria)
#create list of tweet object
text_tweets = [[tweet.date, tweet.username, tweet.text, tweet.retweets, tweet.favorites] for tweet in tweets]
#create dataframe
df_scraping = pd.DataFrame(text_tweets)
df_scraping

```

```
#rename dataframe
df_scraping.rename(columns={0: 'Time', 1: 'Username', 2: 'Text', 3: 'Retweeted', 4:'Liked'}, inplace=True)
df_scraping
```
